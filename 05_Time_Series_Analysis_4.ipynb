# Interview Questions

1. **Question:**
When attempting to make a non-stationary time series stationary using differencing, one must ensure not to over-difference. Over-differencing can lead to which of the following?

- A) Increased predictive accuracy on the training set but reduced accuracy on the test set

- B) A random walk that appears deterministic

- C) A loss of meaningful long-term trend information

- D) The immediate need to apply seasonal ARIMA (SARIMA) models

<br>
<br>
<br>
<br>
<br>
<br>
**Correct Answer:**  C) A loss of meaningful long-term trend information

**Explanation:**  Differencing is used to remove trends and stabilize the mean. However, applying it excessively can remove genuine underlying structure, including trends that may be critical for forecasting.

2. **Question:**
In a seasonal ARIMA (SARIMA) model, the seasonal differencing is primarily intended to handle what feature of the data?

- A) Non-linear anomalies that occur sporadically

- B) Short-term noise fluctuations

- C) Periodic patterns repeating at regular intervals

- D) Trend components that are strictly polynomial

<br>
<br>
<br>
<br>
<br>
<br>
**Correct Answer:**  C) Periodic patterns repeating at regular intervals

**Explanation:**  SARIMA models extend ARIMA by including terms specifically for seasonality. Seasonal differencing targets the repeating patterns that occur at fixed intervals to achieve stationarity.

<br>

**Explanation:**  In exponential smoothing, α determines how fast the influence of older observations decays. A value closer to 1 places greater emphasis on the most recent data point, making the model more reactive but possibly more volatile.

7. **Question:**

```python

# Generate seasonal data
np.random.seed(42)
months = 120  # 10 years of monthly data
time = np.arange(months)
seasonality = 10 * np.sin(2 * np.pi * time / 12)  # Seasonal component
# trend = 0.5 * time  # Trend component
noise = np.random.normal(0, 5, months)  # Noise component
data = seasonality  + noise

# Create a DataFrame
df = pd.DataFrame({"Month": pd.date_range(start="2000-01", periods=months, freq="M"), "Value": data})
df.set_index("Month", inplace=True)

# Apply seasonal differencing (lag 12)
df["Differenced"] = df["Value"].diff(12)

# Plot original data and differenced data
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(df.index, df["Value"], label="Original Data")
plt.title("Original Data with Seasonality")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(df.index, df["Differenced"], label="Seasonally Differenced Data", color="orange")
plt.title("Seasonally Differenced Data (Lag 12)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
mobile_sales = pd.read_excel('mobilesales.xlsx')
# set index
mobile_sales.set_index('DATE', inplace=True)
# linear interpolation
mobile_sales.Sales = mobile_sales.Sales.interpolate(method='linear')
# anomalies - clip quantiles
mobile_sales.Sales = mobile_sales.Sales.clip(upper=mobile_sales.Sales.quantile(0.98), lower=mobile_sales.Sales.quantile(0.02))
# plot
mobile_sales.Sales.plot(style='-o')
```

```python
from sklearn.metrics import (
    mean_squared_error as mse,
    mean_absolute_error as mae,
    mean_absolute_percentage_error as mape

)

# Creating a function to print values of all these metrics.
def performance(actual, predicted):
    print('MAE :', round(mae(actual, predicted), 3))
    print('RMSE :', round(mse(actual, predicted)**0.5, 3))
    print('MAPE:', round(mape(actual, predicted), 3))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

train_max_date = mobile_sales.index[-12]
train_x = mobile_sales.loc[mobile_sales.index < mobile_sales.index[-12]].copy()
test_x = mobile_sales.loc[mobile_sales.index >= mobile_sales.index[-12]].copy()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

train_max_date = mobile_sales.index[-12]
train_x = mobile_sales.loc[mobile_sales.index < mobile_sales.index[-12]].copy()
test_x = mobile_sales.loc[mobile_sales.index >= mobile_sales.index[-12]].copy()

model = SARIMAX(train_x.Sales, order=(3, 1, 3), seasonal_order=(1,1,1,12))
model = model.fit(disp=False)
test_x['pred'] = model.forecast(steps=12)
test_x.plot(style='-o')
performance(test_x['Sales'], test_x['pred'])
```

```python
test_x[['lower', 'upper']] = model.get_forecast(steps=12).conf_int(0.05).values
plt.plot(train_x['Sales'][-20:], '-o', label='training')
plt.plot(test_x['Sales'], '-o', label='actual')
plt.plot(test_x['pred'], '-o', label='prediction')
plt.fill_between(test_x.index, test_x['lower'], test_x['upper'],
                 color='k', alpha=.15)
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
# No need to explain this
def plot_changepoints(signal, changepoints):
  for cp in changepoints:
    plt.axvline(x=cp, color='#FA8072')
  plt.plot(signal, '-*', label='signal')

  start = 0
  trends = np.array([])
  for i in changepoints:
    x = np.arange(start, i)
    y = signal[start:i]
    l1 = np.polyfit(x, y, deg=1)
    trend = x*l1[0] + l1[1]
    trends = np.append(trends, trend)
    start = i

  plt.plot(trends, label='trend')
  plt.legend()
  plt.show()



signal = train_x.Sales # Stationarise for mean as cost function
window=60
def get_slope(y):
  return np.polyfit(np.arange(len(y)), y, deg=1)[0]

changepoints = signal.loc[abs(signal.rolling(window, center=True).apply(get_slope).diff()) > 7.5].index
# converting to index from dates
temp = signal.reset_index()
changepoints = temp.loc[temp['DATE'].isin(changepoints)].index.tolist()

plot_changepoints(signal.values, changepoints)
```
